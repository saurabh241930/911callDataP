```python
import os
train_files = os.listdir("train")
train_files[45]
```

```python
from imageio import imread
import numpy as np

train_data = []

for i,im in enumerate(train_files):
    filename = "train/" + im
    print("on image :" + str(i))
    image = np.asarray(imread(filename))
    train_data.append((image,1 if im.split(".")[0] == "cat" else 0))
```

```python
train_data[0]
```

```python
train_size = int(len(train_files)*0.8)                           #splitting data into 80-20 ratio to TRAIN:TEST
train_data_x = np.zeros((train_size,64,64,3))
train_data_y = np.zeros((1,train_size))                          #creating train data withn 80% of data 
valid_data_x = np.zeros((len(train_files) - train_size,64,64,3)) #creating test data with rest of 20% data 
valid_data_y = np.zeros((1,len(train_files) - train_size))

import warnings
from scipy import misc
warnings.simplefilter("ignore", DeprecationWarning)

for i,(x,y) in enumerate(train_data):
    resized_image = misc.imresize(x,(64,64,3))
    if i < train_size:
        train_data_x[i] = resized_image
        train_data_y[:,i] = y
        print("generating train data : " + str(i))
    else:
        valid_data_x[i - train_size] = resized_image
        valid_data_y[:,i - train_size] = y
        print("generating test data : " + str(i))
```
